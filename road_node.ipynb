```python
import _config
import py_sql
import pandas as pd
import geopandas as gpd
from shapely import wkt

# 데이터 불러오기
df = pd.read_csv(_config.full_path + '.csv', encoding=_config.euc_kr)

# 컬럼 리스트 생성
col_list = []
for col in df.columns:
    if 'geom' in col:
        col_list.append(f"{col} geometry(Geometry, 5170)")
    else:
        col_list.append(f"{col} varchar(100)")

# 테이블 생성 SQL
sql = f'''
    do $$
    begin
        if not exists (select 1 from pg_tables where tablename = 'moct_sgg_2023') then
        create table moct_sgg_2023(
            {', '.join(col_list)}
        );
        end if;
    end $$; 
'''

py_sql.execute_sql(sql)

# geometry 변환
# df['geom'] = df['geom'].apply(wkt.loads)

# 데이터 삽입을 위한 문자열 생성
rows = []
for idx, row in df.iterrows():
    values = []
    for key in row.index:
        if 'geom' in key: values.append(f"'{row[key].wkt}'")
        else: values.append(f"'{row[key]}'")
    rows.append(f"({', '.join(values)})")

# 데이터 삽입 SQL
sql = f'''
    do $$
    begin 
        if exists (select 1 from pg_tables where tablename = 'moct_sgg_2023') then
        insert into moct_sgg_2023 ({', '.join(df.columns)})
        values {', '.join(rows)};
        end if;
    end $$;
'''

print(sql)
py_sql.execute_sql(sql)

```
